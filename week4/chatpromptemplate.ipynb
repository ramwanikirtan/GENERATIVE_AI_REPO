```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_openai import ChatOpenAI 
from dotenv import load_dotenv
load_dotenv()
```

```python
### lets create a chat prompt template

template = ChatPromptTemplate([
    SystemMessage(content = "You are a helpful domain expert assistant in {domain}."),
    HumanMessage(content = "Explain {topic} in 1 line.")
])

## lets fill the template

prompt = template.invoke({'domain': 'AI', 'topic': 'machine learning'})

## lets print the prompt
print(prompt)
```

```python
## the problem is prompt is not filled 
## its very weird behaviour in langchain
## to correct this we will pass the list of tuples


template = ChatPromptTemplate.from_messages([
    ('system', "You are a helpful domain expert assistant in {domain}."),   ## first role type then content
    ('human', "Explain {topic} in 1 line.")
])

prompt = template.invoke({'domain': 'AI', 'topic': 'machine learning'})
print(prompt) 
```
