### Types of LangChain Messages

LangChain provides three main types of messages to structure interactions with language models:

1. HumanMessage: Represents messages sent by the user or human interacting with the language model.
2. AIMessage: Represents responses generated by the AI or language model.
3. SystemMessage: Represents instructions or context provided to guide the behavior of the language model.

These message types help in organizing and managing conversations effectively.

````python
## 1. imports

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from dotenv import load_dotenv
load_dotenv()

## 2. setup the model   
model = ChatOpenAI()
````

````python
## creating messages list
messages=[
    SystemMessage(content= "You are a helpful chat assistant."),
    HumanMessage(content= "Roadmap to learn LangChain?")
]
````

````python
result = model.invoke(messages)
messages.append(AIMessage(content=result.content))
````

````python
print("Messages:", messages)
````
